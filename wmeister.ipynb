```python

x1 = np.linspace(0.0, 5.0, num=25)
x2 = np.linspace(0.0, 5.0, num=15)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)
```

```python
    plt.subplot(1, 1, 1)
plt.plot(x1, y1, 'o-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

plt.subplot(1, 1, 1)
plt.plot(x1, y2, '.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

# plt.show()
```
